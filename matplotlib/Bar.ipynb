# PROGRAMING LANGUAGES COMPARISON

```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df['LanguagesWorkedWith'] = df['LanguagesWorkedWith'].str.split(';')
df = df.explode('LanguagesWorkedWith')
df = df.groupby('LanguagesWorkedWith').count()
df = df.sort_values('Responder_id', ascending = False)
df
```

```python
plt.figure(figsize=(10, 6))

plt.bar(df.index, df.iloc[:, 0])

#plt.xlabel('Programing language')
plt.ylabel('People using language')
plt.title('Most popular programing languages')
plt.xticks(rotation=90)
plt.tight_layout()

#plt.savefig('plot.png')
```

```python
plt.figure(figsize=(10, 6))

plt.barh(df.index, df.iloc[:, 0])

#plt.ylabel('Programing language')
plt.xlabel('People using language')
plt.title('Most popular programing languages')
plt.tight_layout()
```
